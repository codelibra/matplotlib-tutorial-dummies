```python
x = [1,2,3]
y = [5,7,4]

x2 = [11,21,3]
y2 = [0,7,49]

plt.figure(0)
plt.plot(x,y, label="first label")
plt.plot(x2,y2,label="second label")
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("matplotlib tutorial\nCheck it out")
plt.legend()
plt.plot()
```

```python
x = [2,4,6,8,10]
y = [1,5,7,4,15]
plt.figure(1)
plt.bar(x,y, label="bar 1", color='blue')
plt.bar(y,x, label="bar 2", color='red')
plt.xlabel("x")
plt.ylabel("y")
plt.title("matplotlib tutorial\nCheck it out")
plt.legend()
plt.show()
```

```python
days = [1,2,3,4,5]
sleeping = [12,34,16,7,0]
eating  =  [2,1,3,5,6]
working =  [7,8,12,3,2]
playing =  [8,7,4,3,10]

plt.plot([],[],color='m', label="sleeping", linewidth=5)
plt.plot([],[],color='c', label="eating", linewidth=5)
plt.plot([],[],color='r', label="working", linewidth=5)
plt.plot([],[],color='k', label="playing", linewidth=5)

plt.stackplot(days, sleeping, eating,working, playing, colors=['m','c','r','k'])
plt.xlabel("x")
plt.ylabel("y")
plt.title("matplotlib tutorial\nCheck it out")
plt.legend()
plt.show()
```
